```python
import pandas as pd
```

```python
df = pd.read_csv('origin\\final_df.csv')
```

```python
df.drop_duplicates(subset=['text'], keep='first', inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True, how= 'any')
```

```python
df
```

```python
df['less_than_15'] = df['text'].apply(lambda x: len(x) < 15)
```

```python
df
```

```python
df.drop(df[df['less_than_15'] == True].index, inplace=True)
```

```python
df
```

```python
texts = df['text'].tolist()
# cnt = 0
for index, s in enumerate(texts):
    check_text = s[3: 33]
    if check_text is None: continue
    for j in range(index+1, len(texts)):
        try:
            if check_text in texts[j]:
                texts[index] = None
                # print(f'check: {s}')
                # print(f'duplicate at {texts[j]}')
                # print('')
        except:
            print(f'start error: {s}')
            print(f'end error: {texts[j]}')
    # cnt += 1
    # if cnt == 1000:
    #     break
```

```python
df['deduped_text'] = texts
```

```python
df
```

```python
dedup = df.loc[df['deduped_text'].isna()] 
```
