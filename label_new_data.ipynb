# **Sentence Transformer**

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')  # multi-language model
```

```python
sentences_list = ['sao b·∫°n ngu th·∫ø', 'b·∫°n r·∫•t ƒë·∫πp trai', 'b·∫°n r·∫•t x·∫•u trai', 'hom nay tr·ªùi r·∫•t ƒë·∫πp', 'xinh qu√°', 'man d·ª£', "t√¥i v·ª´a tr√∫ng x·ªï s·ªë"]

try_embedding = model.encode(sentences_list, convert_to_tensor= False)
```

```python
cosine_scores = util.cos_sim(try_embedding, try_embedding)

d = {}
for i, v1 in enumerate(sentences_list):
    for j, v2 in enumerate(sentences_list):
        if i >= j:
            continue
        d[v1 + ' vs. ' + v2] = cosine_scores[i][j].item()

# sort by score
d_sorted = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))
d_sorted
```

```python
target_sentences = new_df['comments'].tolist()
```

```python
target_embedding = model.encode(target_sentences, convert_to_tensor=False)
target_embedding.shape
#(6, 384)
```

```python
def knn_with_cos_sim(a: np.array, k: int):
    ## get the index of the k largest values
    indexs = np.argpartition(a, -k)[:, -k:]
    return indexs
    
```

```python
def calculate_similarity(source_embedding, target_embedding, batch_size=5000):
    res = []
    for i in range(0, len(source_embedding), batch_size):
        cosine_scores = util.cos_sim(source_embedding[i:i+batch_size], target_embedding)
        cosine_scores = cosine_scores.cpu().numpy().T
        res.append(knn_with_cos_sim(cosine_scores, k=15).tolist())
    
    return res
```

```python
res = calculate_similarity(source_embedding= embedding, target_embedding= target_embedding, batch_size= 500)
```

# **Get new label**

```python
label_rate = source_df.label.value_counts()[0] / source_df.label.value_counts()[1]
label_rate
```

```python
labels_res = []
for i, labels in enumerate(knn_labels):
    # print(f'source: {new_df['comments'][i]}')
    label = 0
    for j in labels:
        label += source_df['label'][j] * label_rate
        # print(source_df['text'][j])
    label = label / len(labels)
    
    labels_res.append(label)
    # break
```
